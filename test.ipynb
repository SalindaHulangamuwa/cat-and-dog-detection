```python
import os
import numpy as np
from tensorflow.keras.preprocessing.image import img_to_array, load_img
from tensorflow.keras.optimizers import Adam
import tensorflow as tf

# Define data path and categories
data_path = r'C:\Users\ASUS USER\Desktop\New folder (4)\dataset\test'
categories = ['Dogs', 'cats']

# Define labels dictionary
labels_dict = {'Dogs': 0, 'cats': 1}

# Define image dimensions
img_height = 300
img_width = 300

# Load images and labels function
def load_images_and_labels(data_path, categories, labels_dict, img_height, img_width):
    images = []
    labels = []
    for category in categories:
        folder_path = os.path.join(data_path, category)
        for filename in os.listdir(folder_path):
            img_path = os.path.join(folder_path, filename)
            try:
                img = load_img(img_path, target_size=(img_height, img_width))
                img_array = img_to_array(img)
                images.append(img_array)
                labels.append(labels_dict[category])
            except Exception as e:
                print(f"Error loading image '{img_path}': {e}")
    return np.array(images), np.array(labels)

# Load test data
test_images, test_labels = load_images_and_labels(data_path, categories, labels_dict, img_height, img_width)

# Normalize test images
test_images = test_images / 255.0

# Create data generator for testing
test_datagen = tf.keras.preprocessing.image.ImageDataGenerator()

# Create test data generator
test_generator = test_datagen.flow(test_images, test_labels, batch_size=32, shuffle=False)

# Load and compile the best model
best_model = tf.keras.models.load_model("best_model_adam.h5")
best_model.compile(optimizer=Adam(learning_rate=0.0005), loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Evaluate the model on test data
test_loss, test_accuracy = best_model.evaluate(test_generator, steps=len(test_generator), verbose=1)
print(f"Test Loss: {test_loss}")
print(f"Test Accuracy: {test_accuracy}")

# Predict on test data
predictions = best_model.predict(test_generator, steps=len(test_generator), verbose=1)
predicted_classes = np.argmax(predictions, axis=1)

# Print all predictions with true values
for i in range(len(test_labels)):
    true_label = 'cat' if test_labels[i] == 1 else 'dog'
    predicted_label = 'cat' if predicted_classes[i] == 1 else 'dog'
    print(f"Image {i+1}: True Label: {true_label}, Predicted Label: {predicted_label}")
```
